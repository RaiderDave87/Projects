```python
# Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import CSV file
csvpath = "TMDb_Data.csv"
Marvel_df = pd.read_csv(csvpath)
Marvel_df.head()
```

```python
Top_Five_Budget = Marvel_df.sort_values(["Budget"], ascending = False).head()

Films = list(Top_Five_Budget["Films"])
Films_x = np.arange(0, len(Films))

Budget = list(Top_Five_Budget["Budget"])

tick_locations = []
for x in Films_x:
    tick_locations.append(x)

plt.title("Top Five Budgeted Films by Marvel Studios")
plt.xlabel("Films")
plt.ylabel("Budget (M$)")

plt.bar(Films_x, Budget, width = 0.6)
plt.xticks(tick_locations, Films, rotation = 90)
plt.show()
```

```python
Top_Five_Sales = Marvel_df.sort_values(["Box Office Sales (Global)"], ascending = False).head()

Films = list(Top_Five_Sales["Films"])
Films_x = np.arange(0, len(Films))

Sales = list(Top_Five_Sales["Box Office Sales (Global)"])

tick_locations = []
for x in Films_x:
    tick_locations.append(x)

plt.title("Top Five Box Office Sales (Global) Films by Marvel Studios")
plt.xlabel("Films")
plt.ylabel("Box Office Sales (Global)($)")

plt.bar(Films_x, Sales, width = 0.6)
plt.xticks(tick_locations, Films, rotation = 90)
plt.show()
```

```python
Top_Five_Ratings = Marvel_df.sort_values(["Rating"], ascending = False).head()

Films = list(Top_Five_Ratings["Films"])
Films_x = np.arange(0, len(Films))

Sales = list(Top_Five_Ratings["Rating"])

tick_locations = []
for x in Films_x:
    tick_locations.append(x)

plt.title("Top Five Rated Films by Marvel Studios")
plt.xlabel("Films")
plt.ylabel("Rating")

plt.bar(Films_x, Sales, width = 0.6)
plt.xticks(tick_locations, Films, rotation = 90)
plt.show()
```
